y & = & \left\{\begin{array}{ll}
0 & \left(w_{1} x_{1}+w_{2} x_{2} \leqslant \theta\right) \\
1 & \left(w_{1} x_{1}+w_{2} x_{2}>\theta\right)
\end{array}\right.
\end{eqnarray}

```
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = - 0.7
    tmp = np.sum(x * w) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(x * w) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```
NAND(0,0)
```

```
NAND(1,0)
```

```
NAND(1,1)
```

```
NAND(0,1)
```

```
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5,0.5])
    b = -0.2
    tmp = np.sum(x * w) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```
OR(0,0)
```

```
OR(1,1)
```

```
OR(1,0)
```

```
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```
XOR(0,0)
```

```
XOR(1,0)
```

```
XOR(0,1)
```

```
XOR(1,1)
```
